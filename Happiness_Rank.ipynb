```python
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
df1=dfnew.copy()
df1= df1.drop('Country',axis=1)
```

```python

X = df1.drop("Happiness", axis=1)
Y = df1["Happiness"]
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.3, random_state = 10)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lm = LinearRegression()

lm.fit(X_train, y_train)
```

```python
lm.intercept_
```

```python
pred_train = lm.predict(X_train)
pred_test = lm.predict(X_test)

print(pred_train[:3])
print(pred_test[:3])
```

```python
print ('Fit a model X_train, and calculate MSE with Y_train:', np.mean((y_train - lm.predict(X_train)) ** 2))

print ('Fit a model X_test, and calculate MSE with Y_test:', np.mean((y_test - lm.predict(X_test)) ** 2))
```
