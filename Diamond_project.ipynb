```python
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
newdf=df.copy()
newdf.head()
```

```python
newdf['color'] = newdf['depth'].apply(lambda x: float(x))
newdf.head()
```

```python
sns.distplot(newdf['price'])
```

```python
sns.lmplot(x='carat', y='price', data=newdf)
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
sns.scatterplot(x="carat", y="price",data=newdf, ax=ax)
```

```python
sns.barplot(x = 'cut', y = 'price', order=['Fair','Good','Very Good','Ideal','Premium'],  data=newdf)
```

```python
sns.barplot(x = 'color', y = 'price', data=newdf)
```
